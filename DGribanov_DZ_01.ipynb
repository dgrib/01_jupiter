## Задание 1
Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.
 



## Решение задания 1

Precision => max
1. Отбраковка деталей на предприятии Автоваз. Отбраковать явно дефектные детали, которые сомнительные - их можно пропустить и отдать на сборку. Если признали деталь годной и ошиблись, это не так плохо. Хуже потерять время и собрать меньше автомобилей.
2. Отбирать гречневую крупу чистую от крупинок с черной оболочкой. Если убираем крупинку, то она явно должна быть плохой. Если будем убирать слишком много крупинок (подозрительных), то на выходе останется очень мало крупы в целом.
 
Recall => max
1. Задача не пропустить психов в армию, лучше перепроверить здоровых и потратить время. Гораздо страшнее, если человеку с неуравновешенной психикой будет доверено оружие.
2. Отбраковка деталей на предприятии Роллс-Ройс. Некачественная деталь недопустима. Лучше перепроверить сомнительные детали и потерять время, чем потерять имидж. 


## Задание 2
Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?


## Решения задания 2

Нам нужно объединить метрики Precision и Recall каким-то образом, так как неудобно мониторить две метрики. 
Просто среднее брать некорректно, так как среднее от метрик для разных моделей может быть одним и тем же числом. При этом одна из метрик может быть очень низкой.

Например одна модель выдает precision 0.1; recall 1.0. А вторая модель выдает precision 0.55; recall 0.55. Среднее у этих моделей будет одинаковым. Но в первом случае модель выдает очень низкое значение precision 0.1.

Гармоническое среднее (F-мера), выдает низкое значение при низком значении какой-то из метрик (Precision или Recall). F-мера не выдаст хорошее значение, если какая-то из метрик (Precision или Recall) будет плохой.
